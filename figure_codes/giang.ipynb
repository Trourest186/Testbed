```python
file_name_pod = "/root/Future-Lab/data/complete_time/mec/multiple_pod/storage_multiple_pod.csv"
file_name_container = "/root/Future-Lab/data/complete_time/mec/multiple_container/storage_multiple_container.csv"
file_name_process = "/root/Future-Lab/data/complete_time/mec/multiple_process/storage_multiple_process.csv"

file_list = []
file_list.append(file_name_pod)
file_list.append(file_name_container)
file_list.append(file_name_process)

count = 0
for file_name in file_list:
    with open(file_name, 'r') as file:
        for line in file:
            # Chuyển đổi giá trị từ dạng chuỗi sang số nguyên
            value = int(line.strip())
            
            if count == 0:
                # Thêm giá trị vào mảng pod
                pod.append(value)
            if count == 1:
                container.append(value)

            if count == 2:
                process.append(value)\

    count = count +1
            


print(pod)
print(container)
print(process)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# pod
# container
# process

state = ["1 Request", "2 Request", "3 Request"]
fig, ax = plt.subplots(figsize=(10, 6))

# Plot for data
x_pos = np.arange(len(state))

ax.bar(x_pos - 0.2, pod, width=0.2, align='center', alpha=0.9, ecolor='red', capsize=12, label='Multiple Container')

# Plot for data_1
ax.bar(x_pos, container, width=0.2, align='center', alpha=0.9, ecolor='blue', capsize=12, label='Multiple Process')

# Plot for data_2
ax.bar(x_pos + 0.2, process, width=0.2, align='center', alpha=0.9, ecolor='green', capsize=12, label='Multiple Pod')

ax.set_ylabel("Time average to complete a task", fontsize=15)
ax.set_xticks(x_pos)
ax.set_xticklabels(state, rotation=15, fontsize=12)
ax.legend()
ax.tick_params(axis='both', which='major', labelsize=12)
ax.set_ylim(0, 300)

plt.savefig('../pdf/power-states-comparison.pdf', bbox_inches='tight')
plt.show()
```
